```python
#PRACTICAS POOO
# Desafío:
# Transforma este código en una clase llamada Tablero, que tenga los métodos:

# __init__(filas, columnas)

# colocar_obstaculo(fila, columna)

# mostrar()



class Tablero:
    def __init__(self, filas, columnas):
        self.filas = filas
        self.columnas = columnas
        self.mapa = [["0" for _ in range(columnas)] for _ in range(filas)]
    
    def colocar_obstaculo(self, fila, columna):
        self.mapa[fila][columna] = "X"
    
    def mostrar_tablero(self):
        for fila in self.mapa:
            print(" ".join(fila))


t = Tablero(8,8)
t.colocar_obstaculo(6,6)
t.mostrar_tablero()    
```

```python
# Desafío:
# Transforma este código en una clase Laberinto, que tenga los métodos:

# __init__(filas, columnas)

# colocar_entrada(fila, columna)

# colocar_salida(fila, columna)

# mostrar()






class Laberinto:
    
    def __init__(self,filas,columnas):
        self.filas = filas
        self.columnas = columnas
        self.mapa= [["." for _ in range(columnas)] for _ in range(filas)]
    
    def colocar_entrada(self, fila, columna):
        self.mapa[fila][columna] = "E" 
    
    def colocar_salida(self, fila, columna):
        self.mapa[fila][columna] = "S"
    
    def mostrar_tablero(self):
        for fila in self.mapa:
            print(" ".join(fila)) 
         
    

lab= Laberinto(5,5)
lab.colocar_entrada(0,0)
lab.colocar_salida(4,4)
lab.mostrar_tablero()
  
    
```

```python
# Desafío:
# Convierte todo esto en una clase MapaJuego con los siguientes métodos:

# __init__(filas, columnas)

# colocar_inicio_y_fin(inicio, fin)













































































































































































































































# colocar_obstaculos(lista_obstaculos)

# mostrar()

class MapaJuego:
    
    def __init__(self, filas, columnas):
        self.filas = filas
        self.columnas = columnas
        self.mapa= [["." for _ in range(columnas)] for _ in range(filas)]
    
    def colocar_inicio_y_fin(self, inicio, fin):
        self.mapa[inicio[0]][inicio[1]] = "A"
        self.mapa[fin[0]][fin[1]] = "Z"    
        
    def colocar_obstaculos(self, lista_obstaculos):
        for fila, col in lista_obstaculos:
         self.mapa[fila][col] = "X"
         
    def mostrar_tablero(self):
        for fila in self.mapa:
            print(" ".join(fila)) 
            

mapita = MapaJuego(5,5)
mapita.colocar_inicio_y_fin((0,0), (4,4))
mapita.colocar_obstaculos([(3,3)])
mapita.mostrar_tablero()
    
```
